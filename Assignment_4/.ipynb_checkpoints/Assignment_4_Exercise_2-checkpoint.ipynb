```python
from torchvision import datasets, transforms
from torch.utils.data import Subset

# Load dataset and transform
transform = transforms.Compose([
    transforms.Resize((32, 32)),
    transforms.ToTensor(), 
    transforms.Normalize((0.5,), (0.5,))
])
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

train_dataset = Subset(train_dataset,list(range(600)))
test_dataset = Subset(test_dataset,list(range(600)))
```

## Exercise 1

```python
import torch
import torch.nn as nn

# Define VGG11 model
class VGG11(nn.Module):
    def __init__(self):
        super(VGG11, self).__init__()

        # Conv Block 1
        self.conv1 = nn.Conv2d(1, 64, kernel_size=3, stride=1, padding=1)
        self.bn1 = nn.BatchNorm2d(64)
        self.relu1 = nn.ReLU()
        self.maxpool1 = nn.MaxPool2d(kernel_size=2, stride=2)

        # Conv Block 2
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1)
        self.bn2 = nn.BatchNorm2d(128)
        self.relu2 = nn.ReLU()
        self.maxpool2 = nn.MaxPool2d(kernel_size=2, stride=2)

        # Conv Block 3
        self.conv3 = nn.Conv2d(128, 256, kernel_size=3, stride=1, padding=1)
        self.bn3 = nn.BatchNorm2d(256)
        self.relu3 = nn.ReLU()

        # Conv Block 4
        self.conv4 = nn.Conv2d(256, 256, kernel_size=3, stride=1, padding=1)
        self.bn4 = nn.BatchNorm2d(256)
        self.relu4 = nn.ReLU()
        self.maxpool4 = nn.MaxPool2d(kernel_size=2, stride=2)

        # Conv Block 5
        self.conv5 = nn.Conv2d(256, 512, kernel_size=3, stride=1, padding=1)
        self.bn5 = nn.BatchNorm2d(512)
        self.relu5 = nn.ReLU()

        # Conv Block 6
        self.conv6 = nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1)
        self.bn6 = nn.BatchNorm2d(512)
        self.relu6 = nn.ReLU()
        self.maxpool6 = nn.MaxPool2d(kernel_size=2, stride=2)

        # Conv Block 7
        self.conv7 = nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1)
        self.bn7 = nn.BatchNorm2d(512)
        self.relu7 = nn.ReLU()
        
        # Conv Block 8
        self.conv8 = nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1)
        self.bn8 = nn.BatchNorm2d(512)
        self.relu8 = nn.ReLU()
        self.maxpool8 = nn.MaxPool2d(kernel_size=2, stride=2)

        # Fully Connected Layers
        self.fc9 = nn.Linear(512, 4096)
        self.relu9 = nn.ReLU()
        self.dropout9 = nn.Dropout(0.5)

        self.fc10 = nn.Linear(4096, 4096)
        self.relu10 = nn.ReLU()
        self.dropout10 = nn.Dropout(0.5)

        self.fc11 = nn.Linear(4096, 10)

    def forward(self, x):
        x = self.maxpool1(self.relu1(self.bn1(self.conv1(x))))
        x = self.maxpool2(self.relu2(self.bn2(self.conv2(x))))
        x = self.relu3(self.bn3(self.conv3(x)))
        x = self.maxpool4(self.relu4(self.bn4(self.conv4(x))))
        x = self.relu5(self.bn5(self.conv5(x)))
        x = self.maxpool6(self.relu6(self.bn6(self.conv6(x))))
        x = self.relu7(self.bn7(self.conv7(x)))
        x = self.maxpool8(self.relu8(self.bn8(self.conv8(x))))

        # Flatten data
        x = x.view(x.size(0), -1)

        x = self.dropout9(self.relu9(self.fc9(x)))
        x = self.dropout10(self.relu10(self.fc10(x)))
        x = self.fc11(x)

        return x

# Create an instance of the VGG11 model
vgg11_model = VGG11()

# Print the model architecture
print(vgg11_model)
```

## Exercise 2

```python
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import torch.optim as optim 

# Create instance of MLP model
vgg11 = VGG11()

# Define loss criterion as cross entropy loss
criterion = nn.CrossEntropyLoss()

# Define optimizer with learning rate of 0.01 and momentum of 0.9
optimizer = optim.SGD(vgg11.parameters(), lr=0.01, momentum=0.9)

# Use DataLoader to batch and shuffle in each epoch
train_loader = DataLoader(dataset=train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(dataset=test_dataset, batch_size=64, shuffle=False)

# Store losses
train_losses = []
test_losses = []

# Store errors
train_errs = []
test_errs = []

num_epochs = 20

for epoch in range(num_epochs):
    # Set model to training mode
    vgg11.train()
    
    train_loss = 0.0
    
    # Train on training data
    for inputs, labels in train_loader:        
        # Set gradients to 0
        optimizer.zero_grad()

        # Find outputs
        outputs = vgg11(inputs)

        # Calculate loss and update weights
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        train_loss += loss.item()
    
    # Set model to evaluation mode
    vgg11.eval()
    
    test_loss = 0.0
    
    # Compute loss on test data
    for inputs, labels in test_loader:
        # Find outputs
        outputs = vgg11(inputs)

        # Calculate loss
        loss = criterion(outputs, labels)

        test_loss += loss.item()
        
    # Store losses in list
    train_losses.append(train_loss / len(train_loader))
    test_losses.append(test_loss / len(test_loader))
    
    # Compute errors on test data
    test_correct = 0
    test_total = 0

    with torch.no_grad():
        for inputs, labels in test_loader:
            # Find outputs
            outputs = vgg11(inputs)
            
            # Find class with the highest score
            _, predicted = torch.max(outputs.data, 1)
            
            # Calculate accuracy and error
            test_total += labels.size(0)
            test_correct += (predicted == labels).sum().item()
            
    test_acc = test_correct / test_total
    test_err = (1 - test_acc) * 100

    test_errs.append(test_err)
    
    # Compute errors on training data
    train_correct = 0
    train_total = 0

    with torch.no_grad():
        for inputs, labels in train_loader:
            # Find outputs
            outputs = vgg11(inputs)
            
            # Find class with the highest score
            _, predicted = torch.max(outputs.data, 1)
            
            # Calculate accuracy and error
            train_total += labels.size(0)
            train_correct += (predicted == labels).sum().item()
    train_acc = train_correct / train_total
    train_err = (1 - train_acc) * 100

    train_errs.append(train_err)
    
    print(f'Epoch {epoch + 1}/{num_epochs}, Training Loss: {train_loss / len(train_loader):.2f}, Test Loss: {test_loss / len(test_loader):.2f}, Training err: {train_err:.2f}%, Test err: {test_err:.2f}%')
    
plt.plot(train_losses)
plt.plot(test_losses)
plt.legend(['Training loss', 'Test loss'])
plt.title("Loss vs number of epochs")
plt.ylabel("Loss")
plt.xlabel("Number of epochs")
plt.show()

plt.figure()
plt.plot(train_errs)
plt.plot(test_errs)
plt.legend(['Training error', 'Test error'])
plt.title("Error vs number of epochs")
plt.ylabel("Error (%)")
plt.xlabel("Number of epochs")
plt.show()
```
